```python
from tensorflow.keras.applications.vgg16 import VGG16,decode_predictions,preprocess_input
```

```python
model=VGG16()
```

```python
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import load_img,img_to_array
```

```python
img=load_img("d:/images/dog.jpg",target_size=(224,224))
np_img=img_to_array(img)
img_after_preprocess=preprocess_input(np_img.reshape(1,224,224,3))
probs=model.predict(img_after_preprocess)
decode_predictions(probs,top=2)
```
