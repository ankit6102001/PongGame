```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.xception import Xception,preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense,Flatten
```

```python
train_gen=ImageDataGenerator(preprocessing_function=preprocess_input,vertical_flip=True,horizontal_flip=True,zoom_range=.3,rotation_range=.5)
train_itr=train_gen.flow_from_directory("d:/dataset/cat_dog_car/train/",target_size=(224,224),class_mode='binary',batch_size=200)
```

```python
pre_model=Xception(include_top=False,input_shape=(224,224,3))
```

```python
pre_model.summary()
```

```python
join_layer=pre_model.output #returns last layer of pre_model and we will use it as input to our model
fl=Flatten()(join_layer)
d1=Dense(100,activation='relu')(fl)
out_layer=Dense(3,activation='softmax')(d1)

#to disable training of pretrained layers
for layer in pre_model.layers:
    layer.trainable=False

mymodel=Model(inputs=pre_model.input,outputs=out_layer)
mymodel.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
mymodel.fit(train_itr,epochs=2,steps_per_epoch=6000/200)
```
